### Try for all crab files & characterize width ###

```python
W_eff = []

for crab_file in crab_files:
    max_beam = crab_file._max_beam_idx
    time_srs = np.nansum(crab_file.beams[max_beam].intensity,axis=0)
    
    max_int = np.max(time_srs)
    max_idx = np.argmax(time_srs)

    window = np.arange(max_idx - 10, max_idx + 10, 1)

    masked_ts = [time_srs[i] for i in window]
    
    fluence = np.nansum(masked_ts, axis=0)
    w_eff = fluence/max_int
    W_eff.append(w_eff)


print(W_eff)
```

```python
fig, ax = plt.subplots(
    figsize = (8, 8),
    constrained_layout = True)

ax.hist(W_eff, 
        bins = 'auto',
        histtype = 'stepfilled')

ax.set_xlabel("Pulse effective width")
ax.set_ylabel("No. of pulses");
```

### Calculating snr and applying hmk equation ###

```python
## Getting SNRs 
noise_win = np.arange(0, 10, 1)
SNRs = []

for crab_file in crab_files:
    max_beam = crab_file._max_beam_idx
    time_srs = np.nansum(crab_file.beams[max_beam].intensity,axis=0)
    max_int = np.max(time_srs)

    noise_ts = [time_srs[i] for i in noise_win]
    sigma_noise = np.std(noise_ts)
    snr = max_int/sigma_noise
    SNRs.append(snr)

print(SNRs)
```

```python
fig, ax = plt.subplots(
    figsize = (4, 4),
    constrained_layout = True)

ax.hist(SNRs, 
        bins = 'auto',
        histtype = 'stepfilled')

ax.set_title("SNRs for ms timescale")
```

```python
ms_SNR = np.mean(SNRs)
ms_SNR
```

```python
def SNR_ns(SNR_ms, w_ns):
    result = SNR_ms * 10**3 / w_ns
    return result 
```

```python
def SNR_mus(SNR_ms, w_mus):
    result = SNR_ms * np.sqrt(10**3) / w_mus
    return result 
```

```python
## Graphs for a variety of pulse widths ##

my_x = np.arange(0.2, 100, 0.5)
my_y = SNR_ns(SNR_ms = ms_SNR,
              w_ns = my_x)

fig, ax = plt.subplot_mosaic(
    '''
    AC
    BD
    ''',
    figsize = (8, 6),
    constrained_layout = True,
    gridspec_kw ={'wspace': 0.08, 'hspace': 0.04}
)

## Nanosecond graphs ##

ax['A'].set_xlim(0.0, 20)
ax['A'].plot(my_x, my_y,
             color = 'b',
             marker = 'None',
             linestyle = '-',
             lw = 1.2)
#ax['A'].set_xlabel('Pulse width, ns')
ax['A'].set_ylabel('SNR for ns')
ax['A'].set_title('For ns sampling time')
ax['A'].ticklabel_format(axis ='y', style = 'sci', scilimits = (3,3))
ax['A'].grid(True, which='both', color='gray', lw=0.4)

ax['B'].set_xscale('log')
ax['B'].plot(my_x, my_y,
             color = 'b',
             marker = 'None',
             linestyle = '-',
             lw = 1.2)
ax['B'].grid(True, which='both', color='gray', lw = 0.4)
ax['B'].ticklabel_format(axis ='y', style = 'sci', scilimits = (3,3))
ax['B'].set_xlabel('Pulse width, ns')
ax['B'].set_title('semi-log', fontsize = '11')

## Microsecond graphs ## 
my_y = SNR_mus(SNR_ms = ms_SNR,
                   w_mus = my_x)

ax['C'].set_xlim(0.0, 20)
ax['C'].plot(my_x, my_y,
             color = 'b',
             marker = 'None',
             linestyle = '-',
             lw = 1.2)
#ax['C'].set_xlabel('Pulse width, microsec')
ax['C'].set_ylabel('SNR for microsec')
ax['C'].set_title('For microsec sampling time ')
ax['C'].grid(True, which='both', color='gray', lw=0.4)

ax['D'].set_xscale('log')
ax['D'].plot(my_x, my_y,
             color = 'b',
             marker = 'None',
             linestyle = '-',
             lw = 1.2)
ax['D'].grid(True, which='both', color='gray', lw = 0.4)
ax['D'].set_xlabel('Pulse width, microsec')
ax['D'].set_title('semi-log', fontsize = '11')

plt.suptitle('Given SNR = 8 for ms sampling time', fontsize = '15')

plt.show()
```

```python
## Graphs for set pulse width ## 

my_x = np.arange(1, 15, 1) # SNR in ms values 
my_y_1 = SNR_ns(SNR_ms = my_x, 
                w_ns = 0.2)
my_y_2 = SNR_ns(SNR_ms = my_x, 
                w_ns = 2)
my_y_2 = SNR_mus(SNR_ms = my_x, 
                w_mus = 4)

fig, ax = plt.subplot_mosaic(
    '''
    ABC
    ''',
    figsize = (10, 10/3),
    constrained_layout = True
)

ax['A'].plot(my_x, my_y_1,
             color = 'b',
             marker = 'None',
             linestyle = '-',
             lw = 1.2)
ax['A'].grid(True, which='both', color='gray', lw=0.4)
ax['A'].set_xlabel('SNR (ms sampling time)')
ax['A'].set_ylabel('SNR (ns sampling time)')
ax['A'].set_title('0.4 ns pulse\n (Hankins 2007)')
ax['A'].ticklabel_format(axis ='y', style = 'sci', scilimits = (3,3))

ax['B'].plot(my_x, my_y_2,
             color = 'b',
             marker = 'None',
             linestyle = '-',
             lw = 1.2)
ax['B'].grid(True, which='both', color='gray', lw=0.4)
ax['B'].set_xlabel('SNR (ms sampling time)')
ax['B'].set_ylabel('SNR (ns sampling time)')
ax['B'].set_title('2 ns pulse\n (Jessner 2018)')

ax['C'].plot(my_x, my_y_2,
             color = 'b',
             marker = 'None',
             linestyle = '-',
             lw = 1.2)
ax['C'].grid(True, which='both', color='gray', lw=0.4)
ax['C'].set_xlabel('SNR (ms sampling time)')
ax['C'].set_ylabel('SNR (microsec sampling time)')
ax['C'].set_title('4 microsec pulse\n (Hankins 2007)')

plt.suptitle('Expected SNRs for given pulse widths', fontsize = 16)
```
